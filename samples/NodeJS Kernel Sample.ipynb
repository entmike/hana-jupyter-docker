```javascript
let conn = require("@sap/hana-client").createConnection();
```

```javascript
conn.connect({
    serverNode : "hxe:39017",
    uid : "SYSTEM",
    pwd : "HXEHana1"
}, err=>{
    if(err && err.code != -20004){
        console.log(err);
        conn.disconnect();
    }else{
        conn.exec("SELECT TOP 20 * FROM M_TABLES ORDER BY table_size DESC", null, (err, results)=>{
            if(err){
                console.log(err);
            }else{
                console.log(results);
            }
            conn.disconnect();
        });
    }
})
```
